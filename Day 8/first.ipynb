```python
import requests
```

```python
response = requests.get("https://en.wikipedia.org/wiki/Programming_language")
```

```python
from bs4 import BeautifulSoup

text = response.text
parse = BeautifulSoup(text, 'html.parser')
```

```python
print(parse.prettify())
```

```python
headings = parse.find_all('div', class_='mw-heading mw-heading2')
```

```python
heading_text=[]
for heading in headings:
    heading_text.append(heading.text)
```

```python
heading =[text.split('[edit]')[0] for text in heading_text]

heading
    
```
